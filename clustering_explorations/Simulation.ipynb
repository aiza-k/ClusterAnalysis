This file contains a simulation clustering of products by their output (biomass estimates). We generate fake data by using real data and adding noise. We plot the data and analyze the mean, variance and other statistical variables. We then apply multiple clustering algorithms and analyze their output. 

We use a custom function to output EUCLIDEAN distances between clusters and from each data point to it's cluster center. These values are outputted along with the clutsering results. 

At the end of the file, we have also run clustering of the data WITH features from teh product characteristics included. At this point, an appropriate weighting mechanism is needed, since there is far more data than features.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_cluster_product_correlation(data, labels, product_names=None):
    """
    Plot heatmaps of product correlation for products in the same cluster.
    
    Parameters:
    data (np.array or pd.DataFrame): The original data where each row is a product and each column is a feature.
    labels (np.array): Cluster labels for each product.
    product_names (list): Optional list of product names. If not provided, will use index numbers.
    """
    # Convert data to DataFrame if it's not already
    if isinstance(data, np.ndarray):
        data = pd.DataFrame(data)
    
    unique_labels = np.unique(labels)
    
    for label in unique_labels:
        if label == -1:  # Skip noise points if DBSCAN was used
            continue
        
        # Get data for this cluster
        cluster_data = data[labels == label]
        
        # Calculate correlation matrix
        corr_matrix = cluster_data.T.corr()
        
        # Create labels for the heatmap
        if product_names is not None:
            cluster_product_names = [product_names[i] for i in np.where(labels == label)[0]]
        else:
            cluster_product_names = [f"Product {i}" for i in np.where(labels == label)[0]]
        
        # Plot heatmap
        plt.figure(figsize=(12, 10))
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, 
                    xticklabels=cluster_product_names, yticklabels=cluster_product_names)
        plt.title(f'Product Correlation Heatmap for Cluster {label}')
        plt.tight_layout()
        plt.show()

# Example usage:
# Assuming you have already run your clustering algorithm and have data and labels

# data = ... # Your original data
# labels = ... # Cluster labels from your clustering algorithm
# product_names = ... # Optional list of product names

# plot_cluster_product_correlation(data, labels, product_names)
```
